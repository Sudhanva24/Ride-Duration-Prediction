```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("/Users/sudhanvasavyasachi/Desktop/Projects/Hackathon micron/Data/train_data.csv")
test=pd.read_csv("/Users/sudhanvasavyasachi/Desktop/Projects/Hackathon micron/Data/test_data.csv")
val=pd.read_csv("/Users/sudhanvasavyasachi/Desktop/Projects/Hackathon micron/Data/val_data.csv")
```

## Trip distance

```python
train=train[train['trip_distance']>0]
val=val[val['trip_distance']>0]
test=test[test['trip_distance']>0]
```

```python
train.shape
```

## Rate code id

```python
train=train[train['RatecodeID']!=99]
val=val[val['RatecodeID']!=99]
test=test[test['RatecodeID']!=99]
train.shape
```

```python
train=train[train['RatecodeID']!=6]
val=val[val['RatecodeID']!=6]
test=test[test['RatecodeID']!=6]
train.shape
```

```python
train['RatecodeID'].value_counts()
```

## store and forward flag

```python
train.drop(['store_and_fwd_flag'], axis=1, inplace=True)
val.drop(['store_and_fwd_flag'], axis=1, inplace=True)
test.drop(['store_and_fwd_flag'], axis=1, inplace=True)
```

```python
train.head()
```

## Pick up location and drop location

```python
train['route'] = train['PULocationID'].astype(str) + '_' + train['DOLocationID'].astype(str)
val['route'] = val['PULocationID'].astype(str) + '_' + val['DOLocationID'].astype(str)
test['route'] = test['PULocationID'].astype(str) + '_' + test['DOLocationID'].astype(str)
```

```python
train.info()
```

## Tolls

```python
# change toll from bool to 0 and 1
train['tolls'] = train['tolls'].astype(int)
val['tolls'] = val['tolls'].astype(int)
test['tolls'] = test['tolls'].astype(int)
train
```

```python
train.info()
```

```python
# change vendor id to 0 and 1 instead of 1 and 2 , make passenger count int,ratecode id also int,payemnt type also int
train['VendorID'] = train['VendorID'].astype(int)
val['VendorID'] = val['VendorID'].astype(int)
test['VendorID'] = test['VendorID'].astype(int)
train['passenger_count'] = train['passenger_count'].astype(int)
val['passenger_count'] = val['passenger_count'].astype(int)
train['RatecodeID'] = train['RatecodeID'].astype(int)
val['RatecodeID'] = val['RatecodeID'].astype(int)
test['RatecodeID'] = test['RatecodeID'].astype(int)
train['payment_type'] = train['payment_type'].astype(int)
val['payment_type'] = val['payment_type'].astype(int)
test['payment_type'] = test['payment_type'].astype(int)
```

```python
# remove the pulocation and dolocation id
train.drop(['PULocationID'], axis=1, inplace=True)
val.drop(['PULocationID'], axis=1, inplace=True)    
test.drop(['PULocationID'], axis=1, inplace=True)
train.drop(['DOLocationID'], axis=1, inplace=True)
val.drop(['DOLocationID'], axis=1, inplace=True)
test.drop(['DOLocationID'], axis=1, inplace=True)
```

```python
train.info()
```

```python
route_frequencies = train['route'].value_counts()
train['route_frequency'] = train['route'].map(route_frequencies)
val['route_frequency'] = val['route'].map(route_frequencies)
test['route_frequency'] = test['route'].map(route_frequencies)
```

```python
(train['route_frequency']<10).sum()
```

```python
# let us remove the routes which have less than 10 frequency
train = train[train['route_frequency']>=10]
val = val[val['route_frequency']>=10]
test = test[test['route_frequency']>=10]
print(train.shape)
print(val.shape)
print(test.shape)
```

## Target encoding of route variable

```python
# Calculate mean duration for each route from training data
route_means = train.groupby('route')['duration'].mean()

# Apply the encoding to train, validation and test sets
train['route_encoded'] = train['route'].map(route_means)
val['route_encoded'] = val['route'].map(route_means)
test['route_encoded'] = test['route'].map(route_means)

# Drop the original route column
train.drop(['route','route_frequency'], axis=1, inplace=True)
val.drop(['route','route_frequency'], axis=1, inplace=True)
test.drop(['route','route_frequency'], axis=1, inplace=True)
```

```python
train
```

```python
train.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, drop='first')

# Fit and transform train data
cols_to_encode = ['VendorID', 'RatecodeID', 'payment_type']
encoded_train = ohe.fit_transform(train[cols_to_encode])

# Create column names for encoded features - accounting for drop='first'
encoded_feature_names = []
for i, col in enumerate(cols_to_encode):
    # Skip the first category since it's dropped
    unique_vals = ohe.categories_[i][1:]
    encoded_feature_names.extend([f'{col}_{val}' for val in unique_vals])

# Convert encoded arrays to dataframes
encoded_train_df = pd.DataFrame(encoded_train, columns=encoded_feature_names, index=train.index)
encoded_val_df = pd.DataFrame(ohe.transform(val[cols_to_encode]), columns=encoded_feature_names, index=val.index)
encoded_test_df = pd.DataFrame(ohe.transform(test[cols_to_encode]), columns=encoded_feature_names, index=test.index)

# Drop original columns and concatenate encoded columns
train = pd.concat([train.drop(cols_to_encode, axis=1), encoded_train_df], axis=1)
val = pd.concat([val.drop(cols_to_encode, axis=1), encoded_val_df], axis=1)
test = pd.concat([test.drop(cols_to_encode, axis=1), encoded_test_df], axis=1)
```

## Now we will apply cyclic encoding to the time and normal encoding to the pickup day

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, drop='first')

# Fit and transform train data
cols_to_encode = ['pickup_day']
encoded_train = ohe.fit_transform(train[cols_to_encode])

# Create column names for encoded features - accounting for drop='first'
encoded_feature_names = []
for i, col in enumerate(cols_to_encode):
    # Skip the first category since it's dropped
    unique_vals = ohe.categories_[i][1:]
    encoded_feature_names.extend([f'{col}_{val}' for val in unique_vals])

# Convert encoded arrays to dataframes
encoded_train_df = pd.DataFrame(encoded_train, columns=encoded_feature_names, index=train.index)
encoded_val_df = pd.DataFrame(ohe.transform(val[cols_to_encode]), columns=encoded_feature_names, index=val.index)
encoded_test_df = pd.DataFrame(ohe.transform(test[cols_to_encode]), columns=encoded_feature_names, index=test.index)

# Drop original columns and concatenate encoded columns
train = pd.concat([train.drop(cols_to_encode, axis=1), encoded_train_df], axis=1)
val = pd.concat([val.drop(cols_to_encode, axis=1), encoded_val_df], axis=1)
test = pd.concat([test.drop(cols_to_encode, axis=1), encoded_test_df], axis=1)
```

```python
train
```

```python
# Create cyclic features for pickup_hour
train['pickup_hour_sin'] = np.sin(2 * np.pi * train['pickup_hour']/24)
train['pickup_hour_cos'] = np.cos(2 * np.pi * train['pickup_hour']/24)
val['pickup_hour_sin'] = np.sin(2 * np.pi * val['pickup_hour']/24)
val['pickup_hour_cos'] = np.cos(2 * np.pi * val['pickup_hour']/24)
test['pickup_hour_sin'] = np.sin(2 * np.pi * test['pickup_hour']/24)
test['pickup_hour_cos'] = np.cos(2 * np.pi * test['pickup_hour']/24)

# Drop the original pickup_hour column
train.drop(['pickup_hour'], axis=1, inplace=True)
val.drop(['pickup_hour'], axis=1, inplace=True)
test.drop(['pickup_hour'], axis=1, inplace=True)
```

```python
# drop the pickup_hour_cosine
train.drop(['pickup_hour_cos'], axis=1, inplace=True)
val.drop(['pickup_hour_cos'], axis=1, inplace=True)
test.drop(['pickup_hour_cos'], axis=1, inplace=True)
```
